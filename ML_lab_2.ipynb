```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import drive



# Function to explore data
def explore_data(df):
    print("\nDataset Info:")
    print(df.info())
    print("\nStatistical Summary:")
    print(df.describe())
    categorical_cols=df.select_dtypes(include=['object']).columns.tolist()
    numerical_cols=df.select_dtypes(include=['int','float']).columns.tolist()
    print("\nCategorical Columns:",categorical_cols)
    print("Numerical Columns:",numerical_cols)
    return categorical_cols, numerical_cols

# encode categorical attributes
def encode_categorical(df,categorical_cols):
    label_encoders={}
    for col in categorical_cols:
        if df[col].nunique()>2:  # One-Hot Encoding for Nominal
            df=pd.get_dummies(df,columns=[col],drop_first=True)
        else:  # Label Encoding for Ordinal
            le=LabelEncoder()
            df[col]=le.fit_transform(df[col])
            label_encoders[col]=le
    return df,label_encoders

# F handle missing values
def handle_missing_values(df,num_cols):
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if col in num_cols:
                if df[col].skew() > 1:  # Outliers present--> median
                    df[col].fillna(df[col].median(), inplace=True)
                else:  # No outliers-->mean
                    df[col].fillna(df[col].mean(), inplace=True)
            else:  # Categorical--> mode
                df[col].fillna(df[col].mode()[0], inplace=True)
    print("\nMissing Values After Imputation:")
    print(df.isnull().sum())
    return df

# F boxplots for outliers
def plot_outliers(df,numerical_cols):
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df[numerical_cols])
    plt.title("Boxplot for Outliers Detection")
    plt.xticks(rotation=90)
    plt.show()
#Normalization
def normalize_data(df,num_cols):
    scaler=MinMaxScaler()
    df[num_cols]=scaler.fit_transform(df[numerical_cols])
    print("\nNormalized Data:\n", df[num_cols].head())
    return df

#  Jaccard and SMC
def compute_jc_smc(v1,v2):
    p11=np.sum((v1==1)&(v2==1))
    p00=np.sum((v1==0)&(v2==0))
    p01=np.sum((v1==0)&(v2==1))
    p10=np.sum((v1==1)&(v2==0))
    jc=p11/(p01+p10+p11)
    smc=(p11+p00)/(p00+p01+p10+p11)
    return jc,smc

#  cosine similarity
def compute_cosine_similarity(v1, v2):
    return cosine_similarity([v1,v2])[0,1]

# to compute similarity matrices and plot heatmap
def plot_heatmap(df):
    df_subset = df.iloc[:20]
    num_vectors=df_subset.shape[0]
    jc_matrix=np.zeros((num_vectors, num_vectors))
    smc_matrix=np.zeros((num_vectors,num_vectors))
    cos_matrix=np.zeros((num_vectors,num_vectors))

    binary_cols=[col for col in df.columns if set(df[col].unique())=={0, 1}]
    for i in range(num_vectors):
        for j in range(num_vectors):
            jc_matrix[i,j],smc_matrix[i,j]=compute_jc_smc(df_subset.iloc[i][binary_cols],df_subset.iloc[j][binary_cols])
            cos_matrix[i,j]=compute_cosine_similarity(df_subset.iloc[i],df_subset.iloc[j])

    plt.figure(figsize=(10,5))
    sns.heatmap(jc_matrix,annot=True,cmap="coolwarm")
    plt.title("Jaccard Coefficient Heatmap")
    plt.show()
    plt.figure(figsize=(10,5))
    sns.heatmap(smc_matrix,annot=True,cmap="coolwarm")
    plt.title("Simple Matching Coefficient Heatmap")
    plt.show()
    plt.figure(figsize=(10,5))
    sns.heatmap(cos_matrix,annot=True,cmap="coolwarm")
    plt.title("Cosine Similarity Heatmap")
    plt.show()

# Main
drive.mount('/content/drive',force_remount=True)
df=pd.read_excel('/content/Lab Session Data.xlsx',sheet_name='thyroid0387_UCI')
print(df)
categorical_cols,numerical_cols=explore_data(df)
df,_=encode_categorical(df, categorical_cols)
df=handle_missing_values(df, numerical_cols)
plot_outliers(df, numerical_cols)
df=normalize_data(df, numerical_cols)

# similarity measures for first two observations
v1=df.iloc[0].astype(int)
v2=df.iloc[1].astype(int)
jc,smc=compute_jc_smc(v1, v2)
cos_sim=compute_cosine_similarity(v1, v2)

print(f"\nJaccard Coefficient: {jc:.4f}")
print(f"Simple Matching Coefficient: {smc:.4f}")
print(f"Cosine Similarity: {cos_sim:.4f}")

#  heatmap for first 20 observations
plot_heatmap(df)
```
